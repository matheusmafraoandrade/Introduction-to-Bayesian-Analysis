## Exercise 1

```python
# Dados

idades = np.array(range(1,15))
tipos = np.array(['ipe', 'ipe', 'jacaranda', 'manaca'])
subtipos = np.array(['amarelo', 'branco', 'jacaranda', 'manaca'])

df = pd.DataFrame(1, index=np.arange(len(tipos)), columns=idades)
df['tipo'] = tipos
df['subtipo'] = subtipos
df.set_index(['tipo', 'subtipo'], inplace=True)
df
```

### Letra A

```python
total = df.sum().sum()

# Árvores com 1 ano
prob_a = df[1].sum() / total
print(prob_a)

# Ipês amarelos
prob_b = df.iloc[0].sum() / total
print(prob_b)

# Ipês velhos
prob_c = df.iloc[0:2,-3:].sum().sum() / total
print(prob_c)

# Árvores adultas
prob_d = df.iloc[:,3:11].sum().sum() / total
print(prob_d)

# Ipês
prob_e = df.iloc[0:2].sum().sum() / total
print(prob_e)
```

```python
# Dados

intensidades = np.array([1,2,3,4,5,6])

df = pd.DataFrame({'int_1':intensidades})
df = pd.concat([df]*6, ignore_index=True)
df['int_2'] = intensidades.repeat(6)
df['int_total'] = df['int_1'] + df['int_2']
df
```

```python
# Letra A

prob_menor_8 = len(df[(df['int_total'] < 8) & (df['int_1'] < 4)]) / len(df)
prob_menor_8
```

```python
prob_menor_igual_3 = len(df[(df['int_1'] <= 3) | (df['int_2'] <= 3)]) / len(df)
prob_menor_igual_3
```
