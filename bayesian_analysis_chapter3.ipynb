### IMPORTANT! (ONLY FOR DISCRETE CASES):
CDF calculates probability of "at most X". The inverse of this is "greater than X". Thus, to calculate "greater than or equal to X", we do 1-CDF(X-1)

```python
p_c1 = 1/3
p_c2 = 2/3
p_venda_c1 = 1/20
p_venda_c2 = 1/10

p_venda = p_c1*p_venda_c1 + p_c2*p_venda_c2

y = range(0,6) # Dias com venda em um período de 5 dias
dias = 5

dist = stats.binom

dist_py = dist.pmf(y, dias, p_venda, loc=0)
esp = dist.mean(dias, p_venda, loc=0)

print('Distribuição:', dist_py)
print('Valor Esperado:', esp)
```

### Exercise 6

```python
dist = stats.binom
n, p = 84, 0.05 ### ENTENDER O PROBLEMA COMO NÃO COMPARECIMENTOS (FALTAS) DENTRE AS RESERVAS

# Letra A

x = 4 # Para que todos consigam lugar, pelo menos 4 passageiros devem faltar (no máximo 80 devem comparecer)

p_mais_4faltas = 1 - dist.cdf(x-1, n, p) # x-1 pois P(>=4) = 1-P(<4) = 1-P(<=3)
p_mais_4faltas
```

```python
# Letra C

n = 81
prob = 1-dist.cdf(n-81, n, p)

while (prob > 0.8):
  n = n+1
  prob = 1-dist.cdf(n-81, n, p)

print(n, prob)
print(f'No máximo {n-1} bilhetes')
```

### Exercise 17

```python
dist = stats.uniform
a, b = 0, 6

def punif(x):
  return dist.cdf(x, loc=a, scale=b-a)

posterior = ( (punif(2)-punif(0.5)) + (punif(5.5)-punif(4)) ) / (punif(5.5)-punif(0.5))
posterior
```
